```python
%pylab inline 
%matplotlib
import cov19
```

```python
cov_collector = cov19.DataCollector()
```

```python
cases = cov_collector.get_covid19_cases()
countries = ['Canada', 'United_States_of_America', 'Mexico']
cases[countries].plot()
```

```python
countries = ['World']
cases[countries].plot()
```

```python
deaths = cov_collector.get_covid19_deaths()
countries = ['Italy', 'Spain', 'China']
deaths[countries].plot()
```
