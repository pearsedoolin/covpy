```python
%pylab inline
import covpy
```

```python
dataCollector = covpy.DataCollector()
```

```python
cases = dataCollector.get_covid19_cases()
countries = ['Italy', 'China', 'World']
ax = cases[countries].plot()
ax.set_ylabel("Confirmed Covid19 Cases")
```

```python
deaths = dataCollector.get_covid19_deaths()
countries = ['Italy', 'Spain', 'China']
ax = deaths[countries].plot()
ax.set_ylabel("Confirmed Covid19 Deaths")
```
